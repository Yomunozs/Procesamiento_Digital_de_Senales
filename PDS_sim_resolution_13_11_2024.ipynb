```python
# Import needed libraries
import numpy as np
import matplotlib.pyplot as plt
from scipy.fft import fft, ifft, fftshift, fftfreq
```

```python
# Define the parameters for each sine wave
frequencies = [10, 25, 30]  # Frequencies of the sine waves
amplitudes = [1, 0.5, 0.25]  # Amplitudes of the sine waves
duration = 5 # Duration of the signal in seconds
sampling_rate_1 = 100  # Sampling rate in Hz

# Generate time values
time_1 = np.linspace(0, duration, duration * sampling_rate)

# Initialize the signal with zeros
signal_1 = np.zeros(len(time_1))

# Generate each sine wave and add it to the signal
for freq, amp in zip(frequencies, amplitudes):
  wave = amp * np.sin(2 * np.pi * freq * time_1)
  signal_1 += wave

# Plot the signal
plt.figure(figsize=(20, 4))
plt.plot(time_1, signal_1)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sample rate: 100')
plt.grid(True)
plt.show()
```

```python
# Signal with doble points
sampling_rate_2 = 200

# Generate time values
time_2 = np.linspace(0, duration, duration * sampling_rate_2)

# Initialize the signal with zeros
signal_2 = np.zeros(len(time_2))

# Generate each sine wave and add it to the signal
for freq, amp in zip(frequencies, amplitudes):
  wave = amp * np.sin(2 * np.pi * freq * time_2)
  signal_2 += wave

# Plot the signal
plt.figure(figsize=(20, 4))
plt.plot(time_2, signal_2)
plt.xlabel('Time (s)')
plt.ylabel('Amplitude')
plt.title('Sample rate: 200')
plt.grid(True)
```

```python
# Calculate the FFT of y
yf = np.fft.fft(signal_1)
xf = np.fft.fftfreq(len(time_1), time_1[1] - time_1[0])

# Calculate the FFT of y2
yf2 = np.fft.fft(signal_2)
xf2 = np.fft.fftfreq(len(time_2), time_2[1] - time_2[0])

plt.figure(figsize=(20, 6))

# Plot the FFT of y
plt.subplot(1, 2, 1)
plt.plot(xf, np.abs(yf))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT of signal_1')
plt.grid(True)


# Plot the FFT of y2
plt.subplot(1, 2, 2)
plt.plot(xf2, np.abs(yf2))
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT of signal_2')
plt.grid(True)
plt.show()
```

```python
# Perform FFT on both signals
w1 = fft(signal_1)
w2 = fft(signal_2)

# Calculate the magnitude of the FFT
w1_abs = np.abs(w1)
w2_abs = np.abs(w2)

# Calculate the frequencies corresponding to the FFT values
w1_freq = fftfreq(len(w1), 1/100)  # Assuming sampling rate is 100 Hz
w2_freq = fftfreq(len(w2), 1/200)  # Assuming sampling rate is 200 Hz

# Shift the frequencies and FFT values for better visualization
w1_freq_shift = fftshift(w1_freq)
w2_freq_shift = fftshift(w2_freq)
w1_shift = fftshift(w1)
w2_shift = fftshift(w2)

# Visualize signals
plt.figure(figsize=(20, 10))


plt.subplot(2, 4, 1)
plt.plot(w1)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT of signal_1')
plt.grid(True)

plt.subplot(2, 4, 2)
plt.plot(w1_abs)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT(ABS) of signal_1')
plt.grid(True)

plt.subplot(2, 4, 3)
plt.plot(w1_freq, w1_abs)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT + Freq of signal_1')
plt.grid(True)

plt.subplot(2, 4, 4)
plt.plot(w1_freq_shift, w1_shift)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT(Shift) of signal_1')
plt.grid(True)

plt.subplot(2, 4, 5)
plt.plot(w2)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT of signal_2')
plt.grid(True)

plt.subplot(2, 4, 6)
plt.plot(w2_abs)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT(ABS) of signal_2')
plt.grid(True)

plt.subplot(2, 4, 7)
plt.plot(w2_freq, w2_abs)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT + Freq of signal_2')
plt.grid(True)

plt.subplot(2, 4, 8)
plt.plot(w2_freq_shift, w2_shift)
plt.xlabel('Frequency')
plt.ylabel('Magnitude')
plt.title('FFT(Shift) of signal_2')
plt.grid(True)

plt.show()
```
